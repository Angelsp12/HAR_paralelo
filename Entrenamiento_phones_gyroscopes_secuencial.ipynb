```python
# ==============================================================
# BLOQUE 1: Importación de librerías necesarias
# ==============================================================

import pandas as pd
import numpy as np
import os, time
from scipy.fft import fft, fftfreq
from scipy.stats import entropy, mode
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, f1_score, recall_score, precision_score
import seaborn as sns
import matplotlib.pyplot as plt

print(" Librerías importadas correctamente.")
```

```python
# ==============================================================
# BLOQUE 2: Carga y limpieza del dataset (idéntico a la versión paralela)
# ==============================================================

ruta = "/home/samuel/Documentos/datos/Phones_gyroscope.csv"

print("Cargando dataset...")
start_load = time.time()
df = pd.read_csv(ruta, on_bad_lines='skip')
df.dropna(subset=['gt','x','y','z'], inplace=True)
df = df[df['gt'] != 'null']
df['sensor'] = 'gyroscope'
end_load = time.time()

print(f" Dataset cargado con {len(df):,} registros en {end_load - start_load:.2f} s.")

# Codificación de etiquetas y normalización
le = LabelEncoder()
df['activity_code'] = le.fit_transform(df['gt'])
scaler = StandardScaler()
df[['x','y','z']] = scaler.fit_transform(df[['x','y','z']])
print(" Limpieza, codificación y normalización completadas.")
```

```python
# ==============================================================
# BLOQUE 3: Extracción de características (idéntico a la versión paralela)
# ==============================================================

def calculate_sma(window):
    return np.mean(np.abs(window['x']) + np.abs(window['y']) + np.abs(window['z']))

def calculate_spectral_entropy(window):
    entropy_val = 0
    for axis in ['x','y','z']:
        fft_vals = np.abs(fft(window[axis].values))
        fft_power = fft_vals**2
        if np.sum(fft_power) > 0:
            power_dist = fft_power / np.sum(fft_power)
            entropy_val += entropy(power_dist)
    return entropy_val / 3

def extract_features(window):
    features = {}
    for axis in ['x','y','z']:
        features[f'mean_{axis}'] = window[axis].mean()
        features[f'std_{axis}'] = window[axis].std()
    features['corr_xy'] = window['x'].corr(window['y'])
    features['corr_yz'] = window['y'].corr(window['z'])
    features['corr_zx'] = window['z'].corr(window['x'])
    features['sma'] = calculate_sma(window)
    features['spectral_entropy'] = calculate_spectral_entropy(window)
    for axis in ['x','y','z']:
        features[f'jerk_mean_{axis}'] = np.mean(np.diff(window[axis], n=1))
    fft_vals = np.abs(fft(window['x'].values + window['y'].values + window['z'].values))
    fft_freq = fftfreq(len(window), 1/100)
    dom_freq_idx = np.argmax(fft_vals[1:]) + 1
    features['dominant_freq'] = fft_freq[dom_freq_idx]
    for axis in ['x','y','z']:
        features[f'energy_{axis}'] = np.mean(window[axis]**2)
    return pd.Series(features).fillna(0)

print(" Funciones de extracción de características definidas correctamente.")
```

```python
# ==============================================================
# BLOQUE 4: Procesamiento secuencial (sin paralelización)
# ==============================================================

def create_windows_and_extract_features(data_key, device_sensor_df):
    WINDOW_SIZE = 256
    STEP = 128
    if len(device_sensor_df) < WINDOW_SIZE:
        return pd.DataFrame(), np.array([]) 
    
    features_list, labels_list = [], []
    for start in range(0, len(device_sensor_df) - WINDOW_SIZE, STEP):
        end = start + WINDOW_SIZE
        window = device_sensor_df.iloc[start:end]
        features = extract_features(window)
        features['Device'] = data_key[0]
        features_list.append(features)
        labels_list.append(mode(window['activity_code'], keepdims=True).mode[0])
        
    return pd.DataFrame(features_list), np.array(labels_list)

def process_group(group):
    key, data = group
    print(f"Procesando grupo (secuencial): {key} con {len(data):,} registros...")
    features_df, labels_array = create_windows_and_extract_features(key, data)
    if not features_df.empty:
        features_df['label'] = labels_array
    print(f"Grupo {key} finalizado. Se extrajeron {len(features_df):,} ventanas.")
    return features_df

# --- Ejecución SECUENCIAL ---
grouped_data = df.groupby(['Device', 'sensor'])
print(f"\nSe procesarán {len(grouped_data)} grupos de manera SECUENCIAL.\n")

start_feat = time.time()

results_df_list = []
for group in grouped_data:
    results_df_list.append(process_group(group))

end_feat = time.time()

all_features_df = pd.concat(results_df_list, ignore_index=True)

print(f"\n Procesamiento SECUENCIAL completado en {end_feat - start_feat:.2f} s.")
print(f"Dimensiones finales de la matriz de características: {all_features_df.shape}")
```

```python
# ==============================================================
# BLOQUE 5: Entrenamiento y evaluación del clasificador SVM
# ==============================================================

y = all_features_df['label']
X = all_features_df.drop(['label', 'Device'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

svm = SVC(kernel='rbf', C=10, gamma='scale', random_state=42)
start_train = time.time()
svm.fit(X_train, y_train)
end_train = time.time()

print(f"\n Entrenamiento completado en {end_train - start_train:.2f} s.")

# --- Evaluación ---
start_pred = time.time()
y_pred = svm.predict(X_test)
end_pred = time.time()

acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')
rec = recall_score(y_test, y_pred, average='weighted')
prec = precision_score(y_test, y_pred, average='weighted')

print(f"\n Exactitud: {acc*100:.2f}%")
print(f" F1: {f1:.3f} | Recall: {rec:.3f} | Precision: {prec:.3f}")
print(f" Tiempo de predicción: {end_pred - start_pred:.2f} s")

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Matriz de Confusión - Phones Gyroscope (Versión Secuencial)')
plt.xlabel('Predicción')
plt.ylabel('Etiqueta verdadera')
plt.show()
```
